```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
np.random.seed(50)
```

```python
random_2d_arr=np.random.normal(size = [1500,2])
```

```python
x=random_2d_arr[:,:-1]
```

```python
y=random_2d_arr[:,-1]
```

```python
y = 3*x + 8*np.random.normal(size=x.shape)
```

```python
plt.scatter(x,y)
```

# Train test split

```python
x_train=x[:1000]
y_train=y[:1000]
x_train
```

```python
x_test=x[1000:]
y_test=y[1000:]
```

```python
weights = np.random.normal(loc=2.9,size=(10000,2))
```

```python
weights
```

```python
min_error = np.inf
best_i = 0
best_c = 0
for epoch,i in enumerate(weights):
    y_pred = i[0]*x_train + i[1]
    error = mean_squared_error(y_pred,y_train)
    
    if error < min_error:
        best_i = i[0]
        best_c = i[1]
    
```

```python
print(best_i,best_c)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train, best_i*x_train + best_c, color='red', label="Line of Best Fit")
```
