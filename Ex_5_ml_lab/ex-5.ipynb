# Logistic regression

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
```

```python
# Train the Logistic Regression model
from sklearn.preprocessing import StandardScaler
X = data1.iloc[:,:-1].values
Y = data1.iloc[:,-1].values

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)

# Train the Logistic Regression model


# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = LogisticRegression()
model.fit(X_train, Y_train)

```

```python
# Evaluate the model
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

4)
Logistic regression assumes a linear relationship between the independent variables (features) and the log-odds of the dependent variable (churn).
Logistic regression is specifically designed for binary classification tasks, which aligns well with the nature of the Telco-Customer-Churn problem (predicting whether a customer will churn or not). While other algorithms like KNN or decision trees can also handle binary classification, logistic regression might excel in scenarios where the decision boundary is relatively simple and linear.
Logistic regression is generally robust to noise in the data.
As the data to be predicted is <b>binary</b> in nature the best model is logistic regression. Because logistic regression is specifically designed to classify binary output or predictions.
