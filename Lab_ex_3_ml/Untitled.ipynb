# finding number of null values

```python
no_of_nul=list(data.isnull().sum())
nul_columns=list()
```

```python
for i in range(len(no_of_nul)):
    if no_of_nul[i]!=0:
        nul_columns.append(data.columns[i])
```

```python
for i in nul_columns:
    print(i,'                                         ',data[i].dtypes)
```
